## Предобработка данных

### first_day_exposition — тип object вместо datetime

```python
# переводим тип столбец object в datetime
data['first_day_exposition'] = pd.to_datetime(data['first_day_exposition'], format = '%Y-%m-%dT%H:%M:%S')

# смотрим тип столбца
data['first_day_exposition'].head(0)
```

Минимальное значение 1м и максимальное 100м — не реалистичные. Среднее значение 2.77м — реалистичное. Пропуски в столбце высоты потолков могли образоваться, когда это значение не указывалось и подразумевалось обычным. Значит пропущенные значения меняем на медианное.

```python
# присваиваем всем значениям с NaN — медианное значение в столбце ceiling_height
data['ceiling_height'] = data['ceiling_height'].fillna(data['ceiling_height'].median())

# отображаем сводную статистику столбца ceiling_height
data['ceiling_height'].describe()
```

### floors_total — тип float64 вместо int64, количество этажей должно измеряться целыми числами и есть NaN значения

```python
# количество пропущенных значений в floors_total
data['floors_total'].isna().sum()
```

86 пропущенных значений в столбце этажности дома. Пропуски в данном столбце могли образоваться, когда это значение не указывалось и подразумевалось обычным.

```python
# корреляция данных в датафрейме, отсортированном по этажности
data.corr().sort_values(by='floors_total', ascending=False)['floors_total'].reset_index()
```

Значения в столбце коррелируются с количеством комнат. Отрицательной корреляции нет. Значит пропущеные значения меняем на медианное сгруппированное по категории этажей.

```python
# создаем словарь floors_dict с медианными значениями сгруппированными по категории этажей
floors_dict = data.pivot_table(index='floor', values='floors_total', aggfunc='median').reset_index()

# проходим по категориям этажей
for floor in range(data['floor'].min(), (data['floor'].max()+1)):
    data.loc[data['floor'] == floor, 'floors_total'] = (
        data.loc[data['floor'] == floor, 'floors_total']
        .fillna(floors_dict[floors_dict['floor'] == floor]['floors_total'].sum()))
    
# отображаем сводную статистику столбца floors_total
data['floors_total'].describe()
```

Создали словарь с медианными значениями сгруппированными по категории этажей. Прошли циклом по диапазону значений от минимального до максимального значения этажа. Где меняли пропуск в данных на медианное значение из словаря для этого этажа.

```python
#  меняем тип данных в столбце на int
data['floors_total'] = data['floors_total'].astype('int')

# отображаем количество пропущенных значений в floors_total
print(data['floors_total'].isna().sum())

# отображаем тип floors_total
data['floors_total'].dtypes
```

Значения в столбце коррелируются с площадью квартиры и количеству комнат. Отрицательной корреляции не наблюдается. Заполним пропущенные значения по медианному значению общей площади по группам количества комнат.

```python
# создаем столбец доли кухонной площади со значениями жилой площали от общей площади
data['kitchen_ratio'] = data['kitchen_area'] / data['total_area']

# создаем rooms_kitchen_dict с медианными значениями сгруппированное по категории количества комнат
rooms_kitchen_dict = data.pivot_table(index='rooms', values='kitchen_ratio', aggfunc='median').reset_index()

# проходим по категориям количества комнат
for room in range(data['rooms'].min(), (data['rooms'].max()+1)):
    data.loc[data['rooms'] == room, 'kitchen_area'] = (
        data.loc[data['rooms'] == room, 'kitchen_area']
        .fillna(data.loc[data['rooms'] == room, 'total_area'] *
            rooms_kitchen_dict[rooms_kitchen_dict['rooms'] == room]['kitchen_ratio'].sum()))

# обновляю столбцец доли кухонной площади со значениями кухонной площали от общей площади
data['kitchen_ratio'] = data['kitchen_area'] / data['total_area']    
    
# отображаем сводную статистику столбца kitchen_area
data['kitchen_area'].describe()
```

Создали словарь с медианными значениями сгруппированными по категории количества комнат. Прошли циклом по диапазону значений от минимального до максимального значения количества комнат. Где меняли пропуск в данных кухонной площади на медианное значение из словаря для этого количества комнат.

### Уберите редкие и выбивающиеся значения. Опишите, какие особенности обнаружили.

```python
# функция построения списка характерных значений и гистограммы для поиска артифкатов в данных столбца
def find_artifact(frame, column, bins_hist):
  try:
    # расчёт межквартильного размаха
    quart = frame[column].quantile(.75) - frame[column].quantile(.25)

    # расчёт нижней границы
    minim = frame[column].min()
    quart_min = frame[column].quantile(.25) - quart * 4
    if quart_min < minim:
      quart_min = minim
    else:
      quart_min = quart_min
  
    # расчёт верхней границы
    maxim = frame[column].max()
    quart_max = frame[column].quantile(.75) + quart * 4
    if quart_max > maxim:
      quart_max = maxim
    else:
      quart_max = quart_max
    
    # построение гистограммы в полученном диапазоне
    frame.plot(
    title=('hist ' +column), 
    y = column, 
    kind = 'hist', 
    bins = (bins_hist), 
    grid=True, 
    range = ((quart_min), (quart_max)), 
    figsize = (8, 4)
    )
    # отображаем список характерных значений столбца
    display(frame[[column]].describe())
    print('значение медианы между границами гистограммы {}'.format(frame[(frame[column] > quart_min) & (
        frame[column] < quart_max)][column].median()))
    print('начало гистограммы: {}'.format(quart_min))
    print('конец гистограммы: {}'.format(quart_max))
  except:
    print('Ошибка в функции find_artifact')
```
